```julia
using Reactive, Interact, Plots, InteractiveUtils
```

```julia
robot = Robot(0, 0, 0);

steps = 100
dt = 0.1
num_particles = 100

@manipulate for v = 0.1:0.1:2.0, ω = -0.1:0.1:2.0
    plot(xlims=(-10, 10), ylims=(-10, 10), legend=:topleft, xlabel="x", ylabel="y")
    control = Control(v, ω)
    intermediary_results, final_results = montecarlo_sim(robot, control, steps, dt, num_particles)
    plot_control(robot, control)
    plot_results(intermediary_results, final_results)
end
```
